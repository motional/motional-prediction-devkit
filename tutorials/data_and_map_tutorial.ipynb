```python
import pandas as pd
trajectories = pd.read_parquet("sample_data.parquet") #train/2021.05.18.13.20.19_veh-25_03608_03919_00006.parquet
```

```python
trajectories
```

```python
# get all vehicles
vehicles = trajectories[trajectories["category_name"] == "vehicle"]
vehicles
```

```python
# get the trajectory from a vehicle
vid = vehicles["track_token"].unique().tolist()[3]
trajectory = vehicles[vehicles["track_token"]==vid]
trajectory
```

```python
import sys
sys.path.append('../')

# configure your map root, change to your own path
MAP_ROOT = "/data/exp/jinghuan.shang/nuplan/v1.1/maps/"

# get map api for a city
from map.utils import get_all_maps_api, parquet_city_name_to_map_api_name
city = trajectory["city"].unique().tolist()[0]
city = parquet_city_name_to_map_api_name(city)
map_api = get_all_maps_api(MAP_ROOT, [city])[city]
map_api
```

```python
# use this vehicle location as the center of queried map
from map.geometry.geometry import Point2D
vehicle_location = Point2D(trajectory.iloc[0]["x"], trajectory.iloc[0]["y"])

# configure a radius (m)
radius = 20

# use the api
from map.nuplan_map.vector_map_utils import get_neighbor_vector_map
vector_map = get_neighbor_vector_map(map_api, vehicle_location, 20)
```

```python
lane_seg_coords, lane_seg_conns, lane_seg_groupings, lane_seg_lane_ids, lane_seg_roadblock_ids = vector_map
# lane_seg_coords: lane_segment coords in shape of [num_lane_segment, 2, 2].
# lane_seg_conns: lane_segment connections [start_idx, end_idx] in shape of [num_connection, 2].
# lane_seg_groupings: collection of lane_segment indices in each lane in shape of
#     [num_lane, num_lane_segment_in_lane].
# lane_seg_lane_ids: lane ids of segments at given index in coords in shape of [num_lane_segment 1].
# lane_seg_roadblock_ids: roadblock ids of segments at given index in coords in shape of [num_lane_segment 1].
```

```python
# lane segments in within the queired range
# shape of [num_lane_segment, 2, 2]
len(lane_seg_coords), lane_seg_coords[0]
```
